```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.ticker import NullFormatter
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)
```

```python
y = np.random.normal(loc = 0.5, scale = 0.4, size = 1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure(1)
```

```python
# Linear
plt.subplot(221)
plt.plot(x,y)
plt.yscale('linear')
plt.title('lineat')
plt.grid(True)
```

```python
# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)
plt.show()
```

```python
# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)
```

```python
plt.gca().yaxis.set_minor_formatter(NullFormatter())
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
plt.show()
```
